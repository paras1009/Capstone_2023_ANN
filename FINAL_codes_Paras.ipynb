# Code Converter 16-bits (Weights)

```python
def twos_complement(binary_str):
    # Perform two's complement by flipping the bits and adding 1
    binary_str = binary_str.replace('.', '')
    flipped_bits = ''.join('1' if bit == '0' else '0' for bit in binary_str)
    twos_complement = bin(int(flipped_bits, 2) + 1)[2:]
    twos_complement = twos_complement[:6] + '.' + twos_complement[6:]
    return twos_complement.zfill(len(binary_str))

def decimal_to_binary(decimal_num):
    binary_digits = []

    integer_part = abs(int(decimal_num))
    decimal_part = abs(decimal_num) - integer_part

    # Convert integer part to 5-bit binary
    binary_digits = bin(integer_part)[2:].zfill(6)
    binary_digits += "."

    # Convert decimal part to 10-bit binary
    for _ in range(10):
        decimal_part *= 2
        binary_digits += str(int(decimal_part))
        decimal_part -= int(decimal_part)

    # Combine sign bit, integer, and decimal parts
    binary_representation = binary_digits

    if decimal_num < 0:
        # If the number is negative, take two's complement for the entire 16 bits
        binary_representation = twos_complement(binary_representation)

    return binary_representation

# Hardcoded decimal inputs for demonstration (list of lists)
decimal_inputs = [
    [-0.29052734375, -0.00525665283203125],
    [0.199951171875, -0.056243896484375],
    [0.0816650390625, -0.09844970703125],
    [-0.10986328125, -0.88134765625]
]

# Convert each decimal number to binary
binary_results = [[decimal_to_binary(num) for num in sublist] for sublist in decimal_inputs]

print("binary_results: ", binary_results)
```
